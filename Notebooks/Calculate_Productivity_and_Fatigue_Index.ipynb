```python
import pandas as pd

df = pd.read_excel('/home/wissal/Documents/sst_projet.xlsx', sheet_name='Productivity and Fatigue')
```

```python
df = pd.DataFrame(df)

# Calculate productivity and fatigue
df['Productivity'] = (df['Hours worked today'] / 4) * 100
```

```python
import random
import math
from math import log, exp, cos, pi
```

```python
P_baseline_last_day = df['P_baseline_last_day']
#Calculate S_loss for every employee every day 
S_duration = df['S_duration']
```

```python
print(print(df.at[1, 'S_duration']))
```

```python
S_losses, FII = [], []
```

```python
for i in range (0, len(S_duration)):
    S_loss = 8 - S_duration[i]
    S_losses.append(S_loss)
    #Using sleep loss to calculate performance
    P_baseline = 3.9 - 0.4415 * S_loss
    if P_baseline <= P_baseline_last_day[i]: 
        P_cumulative = P_baseline + 0.8720 * (P_baseline_last_day[i]-P_baseline)
    else: 
        P_cumulative = P_baseline + 0.8720 * (P_baseline_last_day[i]-P_baseline)
    #calculate the cumulative component C_f based on performance
    S_p = 1.2484 * log(529*(1.001 - P_cumulative/3.9))
    S_pcc = 1.26 / (1 + exp(3.057 - 0.764 * S_p))
    #calculate the Karolinska Sleepiness Scale
    pKSS = 1.26 / (1 + exp(0.6360 - 1.522 * S_pcc))
    #calculate cumulative fatigue component 
    C_f = 1/3 * pKSS
    print(C_f)
    t_start = 7 #if morning shift
    t_start = 13 #if afternoon shift 
    t_start = random.choice([7, 13])
    #calculate contribution due to start time
    T_start = abs(0.6715 * cos(2*pi / 24 * (t_start - 1.9975)))
    #calculate component of fatigue due to the time of day during shift
    T_timeofday = abs(0.74 * cos(2*pi / 24 * (t_start - 5.23)))
    #calculate shift timing component 
    T_f = 1/3 * (T_start + T_timeofday)
    print(T_f)
    #calculate fatigue : job type/breaks component 
    J_f = 1/3 #with no breaks 
    J_f = 1/5 #else
    J_f = random.choice([1/3,1/5,1/7,1/8,1/6,1/4])
    print(J_f)
    #calculate Fatigue Index 
    FI = 100 * (C_f + J_f + T_f - C_f * (J_f + T_f))
    print(FI)
    FII.append(FI)
```

```python
# Display the resulting table with employee ID, productivity, and fatigue
df['Fatigue'] = FII
result_table = df[['Employee ID', 'Productivity', 'Fatigue']]
result_table
```
