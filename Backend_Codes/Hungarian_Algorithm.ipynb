## rule-based algorithm.

```python
# Creating a predictive model to recommend an employee for a given task

# Function to predict the most suitable employee for a given task
def predict_employee_for_task(task_required_skill, task_day, task_shift, employees_data):
    # Finding employees who have the required skill
    skilled_employees = employees_data[employees_data['Skills'].apply(lambda skills: task_required_skill in skills)]

    # Filter employees who are available on the required day and shift
    available_employees = skilled_employees[
        (skilled_employees['Available_Days'].apply(lambda days: task_day in days)) &
        (skilled_employees['Available_Shifts'].apply(lambda shifts: task_shift in shifts))
    ]

    # If no employees are available, return a message indicating this
    if available_employees.empty:
        return "No suitable employee found"

    # TODO: Further criteria for selecting the best employee can be added here
    # For now, we select the first suitable employee
    selected_employee = available_employees.iloc[0]['Employee_ID']

    return selected_employee

# Example prediction
# Task requires 'Assembly' skill, scheduled for day 4 (Thursday), and during the 1st shift
predicted_employee = predict_employee_for_task('Assembly', 4, 1, employees_data)
predicted_employee
```

## Hungarian Algorithm

```python
import pandas as pd
import numpy as np
from scipy.optimize import linear_sum_assignment


# Creating a cost matrix for the Hungarian Algorithm
# For demonstration, using a random cost matrix
num_employees = len(employees_data)
num_tasks = len(tasks_data)
cost_matrix = np.random.rand(num_tasks, num_employees)

# Applying the Hungarian algorithm
row_ind, col_ind = linear_sum_assignment(cost_matrix)

# Create a DataFrame to display the task to employee assignments
assignments_df = pd.DataFrame({
    'Task_ID': tasks_data['Task_ID'].iloc[row_ind].values,
    'Assigned_Employee_ID': employees_data['Employee_ID'].iloc[col_ind].values
})

assignments_df

```
